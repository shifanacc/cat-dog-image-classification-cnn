# Iris Flower Classification - ANN Model

## Introduction

The Iris flower classification project applies an Artificial Neural Network (ANN) using TensorFlow/Keras to classify Iris flowers into three species: Setosa, Versicolor, and Virginica. The dataset consists of four flower measurements, which are used to train the model to recognize patterns and accurately predict the species of unseen flower samples.


## Objective

The objective of this Project is to develop an Artificial Neural Network (ANN) model using TensorFlow/Keras to accurately classify Iris flowers into Setosa, Versicolor, and Virginica based on their flower measurements.

```python
# Step 1 : Import Libraries

import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Step 2 :Load Dataset

iris = load_iris()

x = iris.data
y = iris.target

# Create DataFrame 
df = pd.DataFrame(x,columns = iris.feature_names)

# Add target column

df['Target'] = y

# Display first Five rows
df.head()

```

```python
# Step 3 : Preprocessing

# Check null values 
print(df.isnull().sum())

# Check Duplicate values
print(df.duplicated().sum())

# Dataset Shape 
print(df.shape)
# Dataset Info
print(df.info())
```

```python
# Step 4 : Separate input and output

x = df.drop('Target',axis = 1)
y = df['Target']
```

```python
# Step 5 : Label Encoding

le = LabelEncoder()

y = le.fit_transform(y)

# One-hot-encoding

y_encoded = to_categorical(y)
print(y_encoded[:5])
```

```python
# Step 6 : Normalize features
scaler = StandardScaler()

X_scaled = scaler.fit_transform(x)
print(X_scaled[:5])
```

```python
# Step 7 :Train-test split

X_train,X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_encoded,
    test_size = 0.2,
    random_state = 42
)

print(X_train.shape)
print(X_test.shape)
```

```python
# Step 8: Build the Model

model = Sequential()
# Hidden Layer 1 
model.add(Dense(10,input_shape=(4,),activation = 'relu'))

# Hidden Layer 2 
model.add(Dense(8,activation='relu'))

# Output Layer 
model.add(Dense(3,activation = 'softmax'))
```

```python
# Step 9 : Compile The model 

model.compile(
    optimizer = 'adam',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
# Step 10 : Train the model

history = model.fit(
    X_train,
    y_train,
    epochs = 50,
    batch_size =8,
    validation_split = 0.2
)
```

```python
# Step : 11 Evaluate the model

loss,accuracy = model.evaluate(X_test, y_test)

print("Loss:",loss)
print("Accuracy",accuracy)
```

```python
# Step 12: Prediction

y_pred = model.predict(X_test)

# Convert predicted values
y_pred_classes = np.argmax(y_pred, axis = 1)

# Convert avtual values
y_test_classes = np.argmax(y_test, axis = 1)

print("Predicted Values:")
print(y_pred_classes[:10])

print("Actual Values :")
print(y_test_classes[:10])
```

```python
# Step 13 : Accuracy Score

from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test_classes,y_pred_classes)
print("Accuracy Score:",acc)
```

```python
# Step 14 : Classification Report

from sklearn.metrics import classification_report

from sklearn.metrics import classification_report

print(classification_report(
    y_test_classes,
    y_pred_classes,
    target_names=["Setosa", "Versicolor", "Virginica"]
))
```

```python
# Step 15 : Save Model

model.save("iris_ann_model.keras")
print("Model saved successfully.")
```

```python
# Step 16 : Save Scaler
import joblib

joblib.dump(scaler, "scaler.pkl")
print("Scaler saved successfully.")
```

```python
# Step 16 : Test with an unseen flower sample
sample = pd.DataFrame({
    'sepal length (cm)': [5.8],
    'sepal width (cm)': [2.7],
    'petal length (cm)': [5.1],
    'petal width (cm)': [1.9]
})
sample_scaled = scaler.transform(sample)
prediction = model.predict(sample_scaled)
predicted_class = np.argmax(prediction)
flower_names = ["Setosa", "Versicolor", "Virginica"]
print("Predicted Flower Species:", flower_names[predicted_class])
```
